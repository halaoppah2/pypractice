```python
import numpy as np

arr = np.array([1, 2, 5, 8, 9])

print(arr)
print(type(arr))
print(arr[0] + arr[1])
print(arr[2:4])

x = arr.copy()
x[0] = 20
print(x)

x = arr.view()
x[1] = 30
print(x)

```

```python
arr = np.array([1, 2, 3])
arr1 = np.array([4, 5, 7])
arr2 = np.concatenate((arr, arr1))
print(arr2)

arr3 = np.array_split(arr2, 3)
print(arr3)
```

```python
arr = np.array([2, 4, 5, 2, 1, 2])
y = np.where(arr == 2)
x = np.where(arr%2 == 0)
z = np.searchsorted(arr, 4)
print(arr)
print(y)
print(x)
print(z)
print(np.sort(arr))
```

```python
import matplotlib.pyplot as plt 
import numpy as np
xpoint = np.array([1, 8])
ypoint = np.array([3, 10])

plt.plot(xpoint, ypoint)
plt.show()
```

```python
xpoint = np.array([1, 8])
ypoint = np.array([3, 10])

plt.plot(xpoint, ypoint, 'o')
plt.show()
```

```python
xpoint = np.array([1, 2, 6, 8])
ypoint = np.array([3, 8, 1, 10])

plt.plot(xpoint, ypoint)
plt.show()
```

```python
ypoint = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoint, marker = 'o')
plt.show()
```

```python
ypoint = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoint, 'o:r', ms = 10, mec = 'r', mfc = 'k')
plt.show()
```

```python
ypoint = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoint, linestyle = 'dashed', color = 'hotpink')
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, marker = 'o', linewidth = '3')
plt.plot(y2)

plt.xlabel("Accommodation Type")
plt.ylabel("Accommodation Count")
plt.title("Cost of Accommodation on UCC Campus")

plt.show()
```
